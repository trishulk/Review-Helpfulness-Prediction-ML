```python
import math
```

```python
cols=event_stats2.columns.tolist()
cols.append("Velocity")
VelInfo = pd.DataFrame(columns=cols)

for i in range(0, len(event_stats2)-1) :
    temp = event_stats2.iloc[i]
    tempnext = event_stats2.iloc[i+1]
    if(temp['Category']=='Fixation' and tempnext['Category']=='Fixation' and (temp['Trial']==tempnext['Trial']) and (temp['Participant']==tempnext['Participant'])) :
        X1 = float(tempnext['Fixation Position X [px]'])
        Y1 = float(temp['Fixation Position Y [px]'])
        X2 = float(tempnext['Fixation Position X [px]'])
        Y2 = float(tempnext['Fixation Position Y [px]'])
        delX = X2-X1
        delY = Y2-Y1
        
        Dist = math.sqrt(delX**2 + delY**2)
        Time = (float(tempnext['Event Start Raw Time [ms]'])-float(temp['Event End Raw Time [ms]']))/1000
        #print("Dist {} ",Dist)
        #print("Time ",Time)
        Velocity = Dist/Time
        temp.set_value('Velocity',Velocity)
        #pd.DataFrame.append(VelInfo,temp)
        VelInfo=VelInfo.append(temp,ignore_index=True)
```
